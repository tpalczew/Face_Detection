```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
from PIL import Image
import numpy as np
```

```python
imagePath = '/Users/tpalczew/SomePhotos/DSC_007.JPG'
```

```python
faceCascade = cv2.CascadeClassifier("/Users/tpalczew/Downloads/FaceDetect-master/haarcascade_frontalface_default.xml")
```

```python
# Read the image
image = cv2.imread(imagePath)
image = cv2.resize(image,(224,224))
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

image.shape
```

```python
def find_faces():
    faces = faceCascade.detectMultiScale(
        gray,
        scaleFactor=1.3,
        minNeighbors=15,
        minSize=(5, 5)   
        )
    faces = faceCascade.detectMultiScale(gray, 1.2, 5)
    print("Found {0} faces!".format(len(faces)))
    
    # Draw a rectangle around the faces
    for (x, y, w, h) in faces:
        print("x = " + str(x))
        print("y = " + str(y))
        print("w = " + str(w))
        print("h = " + str(h))
        cv2.rectangle(image, (x, y), (x+w, y+h), (255, 255, 0), 2)
    
find_faces()
plt.imshow(image)
```
